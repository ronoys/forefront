# Basic Python Pandas Tutorial


Importing Modules

```python
import pandas as pd
from matplotlib import pyplot as plt
```

Access Data Set to Determine Features and Indices

```python
data = pd.read_csv("C:/Users/ronoy/OneDrive/Documents/forefront/season-1819_csv.csv")
data.head()
```

Looking at the data, we can immediately notice all the features and a sample of the data available. For this example we specifically want to focus on matches played by Liverpool. This process is known as **filtering** and can be accomplished using a simple conditional

```python
liverpool = data[(data.AwayTeam == 'Liverpool') | (data.HomeTeam == 'Liverpool')]
```

```python
liverpool.head()
```

This new structure is known as a **series**. It has similar functions to a dataframe that can be used to display simple information about the data.

```python
liverpool.mean()
```

```python
liverpool.corr()
```

To specifically access one feature in a Series, we can create new series which add an attribute if a conditional is met. These statements are in the form:  **newSeries = oldSeries['FeatureToAdd'][conditional]**

```python
goalsScoredHome = liverpool['FTHG'][liverpool['HomeTeam'] == 'Liverpool']
goalsScoredAway = liverpool['FTAG'][liverpool['AwayTeam'] == 'Liverpool']
totalScores = goalsScoredHome.append(goalsScoredAway)
```

```python
totalScores.head()
```

One problem with series is that the indexes are often not in order as they have been taken from a separate data frame. As we can see, the series indeces range from 8 to 303 and are neither in order nor consecutive. To combat this, we can use the *set_index* function

```python
matchList = []
for i in range(0,len(totalScores)):
    matchList.append(i)


liverpool = liverpool.set_index([matchList])
```

```python
liverpool.head()
```

This gives us a complete Series with all Liverpool matches within the past season. In order to plot the goals scores we have to use the *plt.plot()* function. 

```python
plt.plot(totalScores)
```

This correctly plots the goals for each gameweek, but the line is in different orders because of the Home/Away game divide. Using the *sort_index* method we can plot the goals per week in order

```python
plt.plot(totalScores.sort_index())
```

This is just a simple example of data organization and visualization that can be done using matplotlib and pandas. Other modules such as seaborn for visualization and numpy for scientific computing can also be used as alternatives.
